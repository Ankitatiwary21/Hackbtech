```python
import pandas as pd
from textblob import TextBlob
from tkinter import Tk
from tkinter.filedialog import askopenfilename
```

```python
# Function to analyze sentiment
def get_sentiment(text):
    analysis = TextBlob(text)
    if analysis.sentiment.polarity > 0:
        return 'Positive'
    elif analysis.sentiment.polarity < 0:
        return 'Negative'
    else:
        return 'Neutral'
```

```python
import pandas as pd
from textblob import TextBlob
from google.colab import files

# Prompt the user to upload the file
print("Please upload your instadata.csv file for analysis:")
uploaded = files.upload()

# Ensure the file is named 'instadata.csv'
file_name = 'instadata.csv'

# Function to analyze sentiment
def get_sentiment(text):
    analysis = TextBlob(text)
    if analysis.sentiment.polarity > 0:
        return 'Positive'
    elif analysis.sentiment.polarity < 0:
        return 'Negative'
    else:
        return 'Neutral'

# Load the uploaded CSV file
df = pd.read_csv(file_name)

# Apply sentiment analysis to the 'text' column
df['Sentiment'] = df['text'].apply(get_sentiment)

# Display the first few rows with sentiment
print(df.head())

# Save the results to a new CSV file
output_file_name = 'instadata_with_sentiment.csv'
df.to_csv(output_file_name, index=False)

# Provide a download link for the new file
print(f"Sentiment analysis complete. Results saved to: {output_file_name}")
files.download(output_file_name)
```
